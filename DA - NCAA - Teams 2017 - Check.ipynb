```python
import numpy as np
import pandas as pd
```

```python
teams_df = pd.read_csv('/media/sf_QMUL_SD/DataAnalytics/CourseWork/data/teams2017.csv')
print(teams_df.head())
print(teams_df.tail())
```

```python
# create a dummy key
teams_df['key'] = 1

# cross join - multiply each with each (combination) and get only the Id columns
teams_com_df = pd.merge(teams_df, teams_df, how='outer', left_on='key', right_on='key')[['Id_x','Id_y']]

# rename the columns
teams_com_df.columns = ['team_a_id','team_b_id']
teams_com_df
```

```python
# get only the logical combinations:
#  - exclude same teams
#  - exclude the duplicates (higher team Id playing against lower team Id)
#
#  Formula: 68 teams => 68*67 = 2278 possible games
games_df = teams_com_df.loc[teams_com_df.team_a_id < teams_com_df.team_b_id]
games_df
```
